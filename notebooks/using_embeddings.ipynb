```python
import numpy as np

from fence.embeddings.bedrock import TitanMultimodalEmbeddingsG1
from fence.embeddings.bedrock import NovaMultimodalEmbeddings

image_bytes = open("../docs/logo.png", "rb").read()


titan = TitanMultimodalEmbeddingsG1(source="search", output_embedding_length=1024)
text_emb = titan.embed("A fencer with a parrot on their shoulder and the word Fence")
image_emb = titan.embed_image(image_bytes, image_format="png")

# calculate cosine similarity
cosine_similarity = lambda a, b: np.dot(a, b) / (np.linalg.norm(a) * np.linalg.norm(b))

print(f"Similarity: {cosine_similarity(text_emb, image_emb)}")
```
